```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_active = pd.read_csv('cleaned datasets/cleaned-Texas-active-covid-cases.csv')
```

```python
df_active
```

```python
df_active.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
df_active
```

```python
df_active.dropna(inplace = True)
```

```python
df_active
```

```python
df_active.set_index('county_name', inplace = True)
```

```python
df_active
```

```python
df = df_active.T
```

```python
df
```

```python
df.columns
```

```python
counties = list(df.columns)
```

```python
counties
```

```python
df[counties].plot(figsize=(12, 4))
```
