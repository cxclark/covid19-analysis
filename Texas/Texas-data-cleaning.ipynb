## EDA

```python
total_cases_df.head()
```

```python
total_cases_df2 = total_cases_df.copy()
```

```python
total_cases_df2['total'] = total_cases_df.sum(axis = 1)
```

```python
total_cases_df2
```

```python
texas_combined = total_cases_df2[['county_name', 'total']]
```

```python
texas_combined.rename(columns={'total': 'total_cases'}, inplace=True)
```

```python
texas_combined = texas_combined[texas_combined['county_name'] != 'Total']
```

```python
texas_combined
```

```python
total_fatality_df['total'] = total_fatality_df.sum(axis = 1)
total_fatality_df
```

```python
fatalities = total_fatality_df[['County Name', 'total']]
```

```python
fatalities
```

```python
fatalities.rename(columns={'County Name': 'county_name', 'total': 'total_death'}, inplace=True)
```

```python
sentence_case_county = []
for name in fatalities['county_name']:
    name = name.title()
    sentence_case_county.append(name)
```

```python
fatalities['county_name'] = sentence_case_county
```

```python
fatalities
```

```python
texas_combined = pd.merge(texas_combined, fatalities, on='county_name')
```

```python
texas_combined
```
